**Importy bibliotek**

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

**Loading data do dataframes**

```python
df = pd.read_csv("./heart.csv");
```

**Czyszczenie danych**
<p>dane użyte w moim projekcie są 'czyste' pod względem brakujących wartości, wszystkie kolumny zawarte w rozpatrywanym datasecie są wypełnione. </p>
<p>Jedyny outlier w zestawie danych to osoba której ciśnienie krwi (RestingBP) wynosiło 0</p>
<p> usunę więc outliera </p>

```python
df = df.loc[df["RestingBP"] >= 60]
```

```python
df.head()
```

**Changing columns to dummyVariables**

As you can see on the above table, some columns have non-binary and non-numeric values, in order for analysis to be easier to handle in python, I will extract the information to dummyVaraibles columns as shown below:

```python
df = pd.get_dummies(df, columns = ["Sex", "ChestPainType", "RestingECG", 
                              "ExerciseAngina", "ST_Slope"])
```

**Some statistical data about each column**

- count - number of records
- mean - statistical average
- std - standard deviation (**σ**)
- min - minimum
- 25% - lower quartile (**Q1**)
- 50% - median
- 75% - upper quartile (**Q3**)
- max - maximum

```python
df.describe()
```

**Creatign a correlation heatmap**


```python
corr = df.corr()


fig, ax = plt.subplots(figsize=(15,15))
heatmap = sns.heatmap(corr,
           xticklabels=corr.columns,
           yticklabels=corr.columns,
           cmap="Spectral",
           annot= True, ax=ax)

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=75, fontsize=15)
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0, fontsize=15)

heatmap
```
